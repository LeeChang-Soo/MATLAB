###  주가추이, 로그수익률 산포도, 로그수익률 히스토그램

```matlab
price = xlsread('stock_price.xls');

ret = price2ret(price(:,2));

subplot(1, 3, 1)
plot(price(:,2))
title('Stock Path')

subplot(1, 3, 2)
plot(ret, '*')
title('Log-Return plots')

subplot(1, 3, 3)
histfit(ret)
title('Log-Return Histogram')
```

### Wiener Process

```matlab
T  = 1; N=1000; dt = T/N; 
w  = randn(1, N);
dz = w * sqrt(dt);

z = [cumsum(dz)];

plot(z)
xlabel('time')
ylabel('z')
```

### Stock Process(GBM, Euler's Discretization)

```matlab
S=100; mu=0.1; vol=0.3; T=1; N=365; dt = T/N; 
SP = zeros(1, N+1); SP(1) = S;
w  = randn(1, N); 

for i=2:N+1
    SP(i) = SP(i-1) + SP(i-1)*(mu*dt + vol*w(i-1)*sqrt(dt));
end

plot(SP)
xlabel('time')
ylabel('Stock Process')
```
