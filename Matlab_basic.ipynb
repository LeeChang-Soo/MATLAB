```matlab
x = [1 2 3 5 6 7 9 10];
y = [1 2 3 5 4 4 6 9];

p1 = polyfit(x, y, 1);
p2 = polyfit(x, y, 2);
p3 = polyfit(x, y, 3);
p4 = polyfit(x, y, 4);

xp  = 1:1:10;
yp1 = polyval(p1, xp);
yp2 = polyval(p2, xp);
yp3 = polyval(p3, xp);
yp4 = polyval(p4, xp);

%curve fitting plot
subplot(2, 2, 1)
plot(x, y, 'o', xp, yp1)
title('n=1');
subplot(2,2,2)
plot(x, y, 'o', xp, yp2)
title('n=2');
subplot(2,2,3)
plot(x, y, 'o', xp, yp3)
title('n=3');
subplot(2, 2, 4)
plot(x, y, 'o', xp, yp4)
title('n=4');
```

```matlab
X = [1 2 3 7 8 9 11 12];
Y = [1 2 3 5 4 4 6 9 ];
x = 1:0.5:12;

% X, Y 데이터 이용 x데이터에 해당하는 Y값 계산
y_linear = interp1(X, Y, x, 'linear');
y_spline = interp1(X, Y, x, 'spline');

subplot(1, 2, 1)
plot(X, Y, x, y_linear, 'o-')
title('Linear Interpolation')
subplot(1,2,2)
plot(X, Y, x, y_spline, 'o-')
title('Cubic Spline Interpolation')
```
